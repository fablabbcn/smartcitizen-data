```python
import scdata as sc
```

### Device example

```python
d = sc.Device(blueprint='sc_air', params=sc.APIParams(id=17558))
```

```python
# d.load_from_storage()
d.options.limit = 5000
await d.load()
```

```python
# Do not use with "too many points" (i.e. two years of data at 1Min resolution). Recommended maximum is 100000 points
panel = d.ts_panel(frequency='10Min', width=900, height=400)
panel.servable()
```
