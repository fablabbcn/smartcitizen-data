```python
import scdata as sc
```

```python
test = sc.Test(name='EXAMPLE',
            devices=[sc.Device(blueprint='sc_air',
                               params=sc.APIParams(id=id)) for id in [18061, 18062]]
            )
```

```python
for device in test.devices:
    print (device.id)
    test.get_device(device.id).options.limit = 5000
```

```python
await test.load()
```

```python
# Add some booleans to plot. In this case, we are plotting temperature values above the mean of the series
for device in test.devices:
    test.get_device(device.id).data["BOOLEAN"] = test.get_device(device.id).data["TEMP"] > test.get_device(device.id).data["TEMP"].mean()
```

## Plot functions in uPlot

### Controls

- Click to drag
- Zoom inside the plot area by clicking and dragging
- Hold shift to grow - contract

### Single sided plot for test

```python
traces = {
            "1": {"devices": "all",
                    "channel": ["TEMP"],
                    "subplot": 1,
                    "side": "left" # Left side of the plot
                 }
        }

test.ts_uplot(traces = traces,
              formatting = {'width':800,
                            'title': 'Single side plot (left) - (hold shift to grow/contract)',
                            'ylabel': {1: 'Temp [degC]'},  # ylabel or ylabel_left work the same way
                           },
              options = {'frequency': '30Min',
                         'min_date': '2024-10-17', 
                         'max_date': '2024-10-20'})
```

### Two sided plot for test

```python
traces = {
            "1": {"devices": "all",
                    "channel": ["TEMP"],
                    "subplot": 1,
                    "side": "left" # Left side of the plot
                 },
            "2": {"devices": "all",
                    "channel": ["BOOLEAN"],
                    "subplot": 1,
                    "side": "right" # Right side of the plot
                 }
        }

test.ts_uplot(traces = traces,
              formatting = {'width':800,
                            'title': 'Two sided plot - (hold shift to grow/contract)',
                            'ylabel': {1: 'Temp [degC]'},  # ylabel or ylabel_left work the same way
                            'ylabel_right': {1: 'Boolean variable'} # ylabel_right is needed for right side plots
                           },
              options = {'frequency': '1Min',
                         'min_date': '2024-10-17', 
                         'max_date': '2024-10-20'})
```

### Two sided plot for device

```python
traces = {
            "1": {"channel": ["TEMP"],
                    "subplot": 1,
                    "side": "left" # Left side of the plot
                 },
            "2": {"channel": ["BOOLEAN"],
                    "subplot": 1,
                    "side": "right" # Right side of the plot
                 }
        }

test.get_device(18061).ts_uplot(traces = traces,
              formatting = {'width':800,
                            'title': 'Two sided plot - (hold shift to grow/contract)',
                            'ylabel': {1: 'Temp [degC]'},  # ylabel or ylabel_left work the same way
                            'ylabel_right': {1: 'Boolean variable'} # ylabel_right is needed for right side plots
                           },
              options = {'frequency': '1Min',
                         'min_date': '2024-10-17', 
                         'max_date': '2024-10-20'})
```

### Resample and choose format

```python
traces = {
            "1": {"channel": ["TEMP"],
                    "subplot": 1,
                    "side": "left" # Left side of the plot
                 },
            "2": {"channel": ["BOOLEAN"],
                    "subplot": 1,
                    "side": "right" # Right side of the plot
                 }
        }

# Note test.get_device(18061) is a sc.Device
test.get_device(18061).ts_uplot(traces = traces,
              formatting = {'width':800,
                            'title': 'Two sided plot for device - (hold shift to grow/contract)',
                            'ylabel': {1: 'Temp [degC]'},  # ylabel or ylabel_left work the same way
                            'ylabel_right': {1: 'Boolean variable'} # ylabel_right is needed for right side plots
                           },
              options = {'frequency': '10Min', 'resample': 'max', # When resampling, choose min, max or mean (default mean)
                         'min_date': '2024-10-17',
                         'max_date': '2024-10-20'})
```
