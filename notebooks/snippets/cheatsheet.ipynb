### Notes about generalisation

What we generally do is:

- put aside validation set - to check if the data from the training applies for the validation
- put aside test set - to check if you don't overfit to the validation

If you sample randomly for the validation/test set - no need to worry about representativity.

If you use OOB (only for RF) you can skip the validation set somehow, although generally it will be lower than the validation set. This is because the score is calculated based on a subset of the trees, whilst the validation set is using the whole forest.

Using OOB and validation set - gives you probably enough confidence to have a good test.

/!\ Validation set - when there is a temporal component, it is fundamental to use latest data for the validation and test data.

Advice: build 5 models and evaluate them in the validation and test sets. Check that the relationship between both is linear. If not, find a validation set that it's indicative of your test set results. Test variety time wise - 5 models that are trained in the following cases:
- Random
- Last month
- Last 2 weeks...
